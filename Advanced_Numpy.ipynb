```python
filtersize = np.array([[1,0,0],[0,1,0],[0,0,1]])
```

```python
filtersize
```

```python
for k in range(13):
    for j in range(13):
        print(k,"row", (k,k+3), (j,j+3))
```

```python
filSize=3
outputVal=[]
for k in range(13):
    for j in range(13):
        sampD=imgdata[k:k+filSize,j:j+filSize]
        outputVal.append(np.sum(sampD*filtersize))
        print (k,'row ',(k,k+3),(j,j+3))
```

```python
len(outputVal)
```

```python
outputFinal2=np.array(outputVal).reshape(13,13)
```
