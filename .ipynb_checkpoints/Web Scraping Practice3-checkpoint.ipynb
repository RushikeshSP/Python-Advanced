```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.flipkart.com/computers/laptops/pr?sid=6bo%2Cb5g&q=hp+laptops&otracker=categorytree&p%5B%5D=facets.price_range.from%3D75000&p%5B%5D=facets.price_range.to%3DMax"
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")

data = soup.find_all("div", class_ = "_2pi5LC col-12-12")
# print(data)
laptop_name = []
laptop_price = []

for i in data:
    names = i.a.find("div", class_ = "_4rR01T").text if i.find("a", class_="_1fQZEK") else "None"
    laptop_name.append(names)
#     print(names)
    
    price = i.a.find("div", class_ = "_30jeq3 _1_WHN1").text if i.find("a", class_="_1fQZEK") else "0.00"
    laptop_price.append(price)
#     print(price)

# print(laptop_name)
# print(laptop_price)
```

```python
HP_Laptops = pd.DataFrame({"Laptop Name": laptop_name, "Laptop Price": laptop_price})
print(HP_Laptops)
```

```python
HP_Laptops.to_csv("HP_Laptops_Name_and_Price.csv", index = False)
```
