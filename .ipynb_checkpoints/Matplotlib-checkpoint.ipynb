```python
import numpy as np
x = np.array([80,85,90,100,105,110,115,120,125])
y = np.array([40,50,60,70,80,90,100,110,120])

plt.plot(x,y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)

x = np.array([1,1,2,3])
y = np.array([3,6,1,10])

plt.subplot(1,2,2)
plt.plot(x,y)

plt.suptitle("MY SHOP")

plt.show()
```
