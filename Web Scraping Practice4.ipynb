```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.scoopwhoop.com/entertainment/best-shows-on-netflix-india/"
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")

data = soup.find("div", class_ = "article-body")
# print(data)

names = data.find_all("h2")
# print(names)

intro = data.find_all("p", class_="sw-para")
# print(names)

list_of_best_web_series = []

for i in names:
    web_series_name = i.text
#     print(web_series_name)
    list_of_best_web_series.append(web_series_name)

intro_of_best_web_series = []

for i in intro:
    if (i == intro[0] or i == intro[1] or i==intro[52] or i==intro[53]):
        continue
    else:
        web_series_intro = i.text
#         print(web_series_name)
        intro_of_best_web_series.append(web_series_intro)
    
# print(len(list_of_best_web_series))
# print(len(intro_of_best_web_series))
    
```

```python
Best_Web_Series = pd.DataFrame({"Name of web series":list_of_best_web_series,
                                "Description of web series": intro_of_best_web_series})
print(Best_Web_Series)
```

```python
Best_Web_Series.to_csv("Best_Web_Series.csv", index=False)
```
