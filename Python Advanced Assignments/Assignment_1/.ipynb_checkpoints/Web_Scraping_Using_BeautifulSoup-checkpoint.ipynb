```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://docs.python.org/3/library/random.html"
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")

data1 = soup.find_all("dl", class_="function")
data2 = soup.find_all("dl", class_="class")
data = data1+data2
# print(data)

Function_Names = []
Function_Description = []

for func in data:
    function_name = func.find("dt").get_text() 
#     print(function_name)
    Function_Names.append(function_name)
    
    function_use = func.find("dd").get_text() 
#     print(function_use)
    Function_Description.append(function_use)
    
# print(Function_Names)
# print(Function_Description)
    
```

```python
Random_Module_Functions = pd.DataFrame({"Function Name": Function_Names, 
                                        "Function Description": Function_Description})
print(Random_Module_Functions)
```

```python
Random_Module_Functions.to_csv("Random_Module_Functions.csv", index=False)
```
