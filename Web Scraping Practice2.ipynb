```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

movie_name = [] 
year = []
time = []
rating = []
metascore = []
votes = []
gross = []

```python
url = "https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating"
r = requests.get(url)
soup = BeautifulSoup(r.content, "html.parser")

movie_name = [] 
year = [] 
time = [] 
rating = [] 
metascore = [] 
votes = [] 
gross = []

movie_data = soup.find_all("div", class_ = "lister-item mode-advanced")


for i in movie_data:
    name = i.h3.a.text
    movie_name.append(name)
#     print(name)
    
    years = i.h3.find("span", class_="lister-item-year text-muted unbold").text.replace("(", " ").replace(")", " ")
    year.append(years)
#     print(years)
    
    run_time = i.p.find("span", class_="runtime").text
    time.append(run_time)
#     print(run_time)
    
    ratings = i.find("div", class_="inline-block ratings-imdb-rating").text.replace("\n", "")
    rating.append(ratings)
#     print(ratings)

    metascores = i.find("div", class_="inline-block ratings-metascore").text.replace("Metascore", "").replace("\n", "") if i.find("span", class_="metascore") else "None" 
    metascore.append(metascores)
#     print(metascores)

    votes_and_gross = i.find_all("span", attrs = {"name": "nv"})
    
    votes_ = votes_and_gross[0].text
    votes.append(votes_)
#     print(votes_)
    
    gross_ = votes_and_gross[1].text if len(votes_and_gross)>1 else "None"
    gross.append(gross_)
#     print(gross_)
    
```

```python
IMDb_rated_top_100_movies = pd.DataFrame({"Movie Name": movie_name, 
                                         "Year of Release": year, 
                                         "Runtime": time, 
                                         "Ratings": rating, 
                                         "Metascore": metascore, 
                                         "Votes": votes, 
                                         "Gross": gross})
print(IMDb_rated_top_100_movies)
```

```python
IMDb_rated_top_100_movies.to_csv("IMDb_rated_top_100_movies.csv")
```
